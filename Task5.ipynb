```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read CSV File
df = pd.read_csv("sales.csv")
print("CSV Loaded Successfully!\n")

# 2. Show first 5 rows
print("First 5 rows:")
print(df.head())

# 3. Basic info
print("\nData Info:")
print(df.info())

# 4. Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# 5. Total sales
print("\nTotal Sales:")
print(df['Sales'].sum())

# 6. Product-wise total sales
product_sales = df.groupby('Product')['Sales'].sum()
print("\nProduct-wise Sales:")
print(product_sales)

# 7. Plot Product-wise Sales
product_sales.plot(kind='bar')
plt.title("Product-wise Total Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()

# 8. Month-wise sales
monthly_sales = df.groupby('Month')['Sales'].sum()
print("\nMonth-wise Sales:")
print(monthly_sales)

# 9. Plot month-wise sales
monthly_sales.plot(kind='line', marker='o')
plt.title("Month-wise Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
